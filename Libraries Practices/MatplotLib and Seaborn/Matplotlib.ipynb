```python
x = np.arange(0,5*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("Sine Graph")

plt.subplot(2,1,2)
plt.plot(x,y1,'r')
plt.title("Cosine Graph")
```
